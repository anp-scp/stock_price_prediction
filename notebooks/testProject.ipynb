```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfTrain = pd.read_csv("train.csv",header=0)
dfTest = pd.read_csv("test.csv",header=0)
yTest = pd.read_csv("sample_submission.csv",header=0)
```

```python
dfTrain.head(2)
```

```python
dfTrain = dfTrain.set_index("Id")
dfTest = dfTest.set_index("Id")
yTest = yTest.set_index("Id")
dfTrain.head()
```

```python
for col in dfTrain.columns[dfTrain.isna().any()].tolist():
    if(dfTrain[col].dtype == "object"):
        dfTrain[col].fillna("unknown", inplace=True)
    if((dfTrain[col].dtype == "float64") or (dfTrain[col].dtype == "int64")):
        dfTrain[col].fillna(dfTrain[col].mean(), inplace=True)
for col in dfTest.columns[dfTest.isna().any()].tolist():
    if(dfTest[col].dtype == "object"):
        dfTest[col].fillna("unknown", inplace=True)
    if((dfTest[col].dtype == "float64") or (dfTest[col].dtype == "int64")):
        dfTest[col].fillna(dfTest[col].mean(), inplace=True)
```

```python
dfTrain.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for col in dfTest:
    if(dfTest[col].dtype == "object"):
        data = dfTrain[col].append(dfTest[col])
        encoder.fit(data.values)
        dfTrain[col] = encoder.transform(dfTrain[col])
        dfTest[col] = encoder.transform(dfTest[col])
```
